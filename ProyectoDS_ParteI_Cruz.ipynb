## ABSTRACT

### Resumen

Este proyecto se centra en analizar la prevalencia y las causas de la diabetes en Estados Unidos, utilizando un conjunto de datos equilibrado basado en la encuesta BRFSS 2015 del CDC, disponible en Kaggle. La diabetes, una enfermedad crónica que impacta significativamente la salud pública y la economía, afecta a millones de personas en el país. En la primera fase, se busca analizar y visualizar hallazgos clave en los datos de más de 70,000 participantes, lo que permitirá en fases posteriores desarrollar modelos predictivos para mejorar el diagnóstico y tratamiento de la diabetes.

Dataset:
    Teboul, A. (2019). Diabetes Health Indicators Dataset. Kaggle. https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset



### Hipótesis de interés:
#### Basándonos en un análisis previo del conjunto de datos, planteamos las siguientes hipótesis:
- La prevalencia de diabetes varía según el nivel educativo.
- La edad y el género influyen en el padecimiento de la diabetes.
- Factores como la obesidad y la hipertensión aumentan el riesgo de diabetes.
- El consumo del alcohol y tabaco influyen en el padecimiento de la diabetes.


### Objetivos:
- Probar las hipótesis mediante análisis estadísticos y visualizaciones.
- Identificar hallazgos importantes en los datos mediante visualizaciones.
- Mostrar la información de forma fácil y sencilla de comprender para la población  .

### Impacto:
#### Los resultados permitirán:
- Comprender mejor los factores que influyen en el riesgo de diabetes.
- Mejorar las estrategias de prevención y políticas de salud pública para la diabetes.

```python
#Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importación de df

Path = "https://raw.githubusercontent.com/AdoCrz/ProyectoDS_ParteI_Cruz/main/Df_diabetes.csv"
df = pd.read_csv(Path)
```

```python
# Información de la estructura del df
df.info()
```

```python
# Información de los Null
df.isnull().sum()
```

```python
# Transformando los datos a entero (ya que ningún dato contiene números decimales > 0)
df = df.astype('int')
```

```python
# Creando las funciones para crear columnas categóricas que ayuden con la lectura de las gráficas y sea más fácil para los usuarios

def status(x):
    if x['Diabetes_binary'] == 0:
        return "No Diabetes"
    else:
        return "Diabetes"
# ----------------------------------------------------

def hip(x):
    if x['HighBP'] == 0:
        return "No Hipertenso"
    else:
        return "Hipertenso"
# ----------------------------------------------------

def col(x):
    if x['HighChol'] == 0:
        return "No Colesterol alto"
    else:
        return "Colesterol alto"
# ----------------------------------------------------

def Sm(x):
    if x['Smoker'] == 0:
        return "No Fumador"
    else:
        return "Fumador"
# ----------------------------------------------------

def Sx(x):
    if x['Sex'] == 0:
        return "Femenino"
    else:
        return "Masculino"
# ----------------------------------------------------

def af(x):
    if x['PhysActivity'] == 0:
        return "Actividad física inactiva"
    else:
        return "Actividad física activa"
# ----------------------------------------------------

def fr(x):
    if x['Fruits'] == 0:
        return "No consume frutas"
    else:
        return "Consume frutas"
# ----------------------------------------------------

def vg(x):
    if x['Veggies'] == 0:
        return "No consume vegetales"
    else:
        return "Consume vegetales"
# ----------------------------------------------------

def al(x):
    if x['AnyHealthcare'] == 0:
        return "No tiene atención médica"
    else:
        return "Tiene atención médica"
# ----------------------------------------------------

def education(x):
    if x['Education'] == 1:
        return "No se graduó de la escuela secundaria"
    elif x['Education'] == 2:
        return "Graduado de la escuela secundaria"
    elif x['Education'] == 3:
        return "Asistió a la universidad"
    elif x['Education'] == 4:
        return "Graduado universitario"
    elif x['Education'] == 5:
        return "Estudios de posgrado"
    elif x['Education'] == 6:
        return "Doctorado o equivalente"
# ----------------------------------------------------

def age(x):
    if x['Age'] == 1:
        return "18-24 años"
    elif x['Age'] == 2:
        return "25-29 años"
    elif x['Age'] == 3:
        return "30-34 años"
    elif x['Age'] == 4:
        return "35-39 años"
    elif x['Age'] == 5:
        return "40-44 años"
    elif x['Age'] == 6:
        return "45-49 años"
    elif x['Age'] == 7:
        return "50-54 años"
    elif x['Age'] == 8:
        return "55-59 años"
    elif x['Age'] == 9:
        return "60-64 años"
    elif x['Age'] == 10:
        return "65-69 años"
    elif x['Age'] == 11:
        return "70-74 años"
    elif x['Age'] == 12:
        return "75-79 años"
    elif x['Age'] == 13:
        return "80 o más años"
```

```python
# Creando las variables categóricas

df['Status'] = df.apply(status, axis = 1)

df['Hipertenso'] = df.apply(hip, axis = 1)

df['Nivel_Col'] = df.apply(col, axis = 1)

df['Género'] = df.apply(Sx, axis = 1)

df['Actividad_física'] = df.apply(af, axis = 1)

df['Consumo_Frutas'] = df.apply(af, axis = 1)

df['Consumo_vegetales'] = df.apply(fr, axis = 1)

df['Seguro_med'] = df.apply(al, axis = 1)

df['Educación'] = df.apply(education, axis = 1)

df['Edad'] = df.apply(age, axis = 1)
```

```python
# Visualizando las nuevas variables
df[['Status', 'Hipertenso', 'Nivel_Col', 'Género', 'Actividad_física',
       'Consumo_Frutas', 'Consumo_vegetales', 'Seguro_med',
       'Edad', 'Educación']].head()
```

## Análisis gráfico

```python
# Gráficas para analizar la primer y segunda hipótesis
import matplotlib.pyplot as plt

# Configuración de los gráficos
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

# Prevalencia de diabetes según la edad
age_diabetes = df.groupby('age_category')['Diabetes_binary'].mean()
axes[0].bar(age_diabetes.index, age_diabetes.values, color='#265C4B')
axes[0].set_title('Prevalencia de Diabetes según la Edad')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Proporción de Diabetes')
axes[0].tick_params(axis='x', rotation=45)

for i, v in enumerate(age_diabetes.values):
    axes[0].text(i, v + 0.01, f"{v:.2f}", ha='center')

# Distribución de personas con diabetes por nivel educativo
df_no_diabetes = df[df['Diabetes_binary'] == 0]['Education']
df_diabetes = df[df['Diabetes_binary'] == 1]['Education']

axes[1].hist([df_no_diabetes, df_diabetes], bins=6, stacked=True, color=['#265C4B', '#9FC131'], label=['No Diabetes', 'Diabetes'], edgecolor='black')
axes[1].set_title('Distribución del Nivel Educativo')
axes[1].set_xlabel('Nivel Educativo')
axes[1].set_ylabel('Frecuencia')
axes[1].legend()

# Prevalencia de diabetes según la educación
education_diabetes = df.groupby('education_category')['Diabetes_binary'].mean()
axes[2].bar(education_diabetes.index, education_diabetes.values, color='#9FC131')
axes[2].set_title('Prevalencia de Diabetes según la Educación')
axes[2].set_xlabel('Nivel Educativo')
axes[2].set_ylabel('Proporción de Diabetes')
axes[2].tick_params(axis='x', rotation=45)

for i, v in enumerate(education_diabetes.values):
    axes[2].text(i, v + 0.01, f"{v:.2f}", ha='center')

plt.tight_layout()
plt.show()

```

### Interpretación Relacionada con las Primeras Hipótesis

**Prevalencia de Diabetes según la Edad**

La prevalencia de diabetes aumenta significativamente con la edad, confirmando la hipótesis de que la edad influye en la probabilidad de desarrollar diabetes. La proporción pasa de un 8-10% en edades jóvenes (18-29 años) a más del 60% en los grupos mayores (65-79 años).

**Distribución del Nivel Educativo**

Aunque la mayoría de los participantes con un nivel educativo más alto (5 y 6) no tienen diabetes, hay una cantidad considerable en estos grupos que sí la tienen, lo que sugiere una posible influencia del nivel educativo en la prevalencia de diabetes.

**Prevalencia de Diabetes según la Educación**

La educación muestra un patrón interesante; pues la mayor prevalencia se observa en quienes asistieron a la universidad pero no se graduaron (67%) y en graduados de la escuela secundaria (72%), lo que podría indicar que ciertos niveles educativos están más relacionados con un mayor riesgo de diabetes.


```python
#Gráficas para la tercer hipótesis
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

# Distribución del IMC
sns.histplot(df['BMI'], kde=True, color='#265C4B', ax=axes[0])
axes[0].set_title('Distribución del IMC')
axes[0].set_xlabel('Índice de Masa Corporal (IMC)')
axes[0].set_ylabel('Frecuencia')

# Relación entre IMC y Diabetes
sns.boxplot(x='Status', y='BMI', data=df, palette=['#03A63C', '#9FC131'], ax=axes[1])
axes[1].set_title('Relación entre IMC y Diabetes')
axes[1].set_xlabel('Diabetes')
axes[1].set_ylabel('Índice de Masa Corporal (IMC)')

# Relación entre Hipertensión y Diabetes
sns.countplot(x='Hipertenso', hue='Diabetes_binary', data=df, palette=['#03A63C', '#9FC131'], ax=axes[2])
axes[2].set_title('Hipertensión y Diabetes')
axes[2].set_xlabel('Hipertensión')
axes[2].set_ylabel('Frecuencia')
axes[2].legend(title='Diabetes')

plt.tight_layout()
plt.show()
```

### Interpretación Relacionada con la tercer Hipótesis
**Distribución del IMC**

La mayoría de los participantes tienen un IMC entre 20 y 35, indicando una prevalencia significativa de sobrepeso y obesidad en la muestra.

**Relación entre IMC y Diabetes**

Las personas con diabetes tienden a tener un IMC más alto en comparación con quienes no tienen diabetes, apoyando la hipótesis de que la obesidad aumenta el riesgo de diabetes.

**Hipertensión y Diabetes**

La hipertensión es mucho más común en personas con diabetes, lo que respalda la hipótesis de que la hipertensión es un factor de riesgo importante para la diabetes.


```python
#Comparación de Consumo de Tabaco y Diabetes

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

sns.barplot(x='Smoker', y='Diabetes_binary', data=df, estimator=lambda x: sum(x) / len(x), palette=['#265C4B', '#9FC131'], ax=axes[0])
axes[0].set_title('Prevalencia de Diabetes según Consumo de Tabaco')
axes[0].set_xlabel('Consumo de Tabaco (0 = No, 1 = Sí)')
axes[0].set_ylabel('Proporción de Diabetes')

#Comparación de Consumo de Alcohol y Diabetes
sns.barplot(x='HvyAlcoholConsump', y='Diabetes_binary', data=df, estimator=lambda x: sum(x) / len(x), palette=['#03A63C', '#9FC131'], ax=axes[1])
axes[1].set_title('Prevalencia de Diabetes según Consumo de Alcohol')
axes[1].set_xlabel('Consumo de Alcohol (0 = No, 1 = Sí)')
axes[1].set_ylabel('Proporción de Diabetes')

#Comparación de Consumo de Tabaco, Alcohol y Diabetes
sns.heatmap(df.pivot_table(values='Diabetes_binary', index='Smoker', columns='HvyAlcoholConsump', aggfunc='mean'), annot=True, cmap='Greens', ax=axes[2])
axes[2].set_title('Interacción entre Tabaco, Alcohol y Diabetes')
axes[2].set_xlabel('Consumo de Alcohol (0 = No, 1 = Sí)')
axes[2].set_ylabel('Consumo de Tabaco (0 = No, 1 = Sí)')

plt.tight_layout()
plt.show()
```

### Interpretación Relacionada con la cuarta Hipótesis
**Prevalencia de Diabetes según Consumo de Tabaco**

El consumo de tabaco parece estar asociado con un mayor riesgo de diabetes, ya que los fumadores tienen una mayor proporción de diabetes en comparación con los no fumadores.

**Prevalencia de Diabetes según Consumo de Alcohol**

Curiosamente, aquellos que no consumen alcohol en exceso tienen una mayor prevalencia de diabetes en comparación con los consumidores, lo que podría sugerir un patrón de comportamiento complejo en la relación entre el consumo de alcohol y la diabetes.

**Interacción entre Tabaco, Alcohol y Diabetes**

La combinación de tabaco y alcohol muestra que los fumadores que no consumen alcohol tienen la mayor prevalencia de diabetes, lo que indica una posible interacción entre estos factores de riesgo.
